```python
missing_values = Churn.isnull().sum()
```

```python
#Convert 'TotalCharges' to numeric (it may contain non-numeric values)
Churn['TotalCharges'] = pd.to_numeric(Churn['TotalCharges'], errors='coerce')
```

```python
#Standardize categorical values ("No phone service" should be "No")
replace_dict = {
    "No phone service": "No",
    "No internet service": "No"
}
Churn.replace(replace_dict, inplace=True)
```

```python
Churn.drop_duplicates(inplace=True)
```

```python
#Fill missing values in 'TotalCharges' with the mean of the column
Churn['TotalCharges'].fillna(Churn['TotalCharges'].mean(), inplace=True)
```

```python
print(Churn.isnull().sum())
```

```python
#Save the cleaned dataset to a new Excel file
cleaned_file_path = "cleaned_churn_dataset.xlsx"
Churn.to_excel(cleaned_file_path, index=False)
```
